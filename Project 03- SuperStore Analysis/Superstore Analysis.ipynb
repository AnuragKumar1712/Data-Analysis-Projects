# Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load and Check the DataSet

```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

## Drop a Column Temporarily

```python
df.drop(columns='Postal Code')
```

## Drop a Column Permanently

```python
df.drop(columns='Postal Code', inplace=True)
df.head()
```

## Get Unique Values of Ship Mode, Segment, Country and Category

```python
print(df['Ship Mode'].unique())
print(df['Segment'].unique())
print(df['Country'].unique())
print(df['Category'].unique())
```

## Get Statistical information of only numeric columns

```python
df.describe()
```

## Get more Information about the dataset

```python
df.info()
```

## Check for any missing Values

```python
df.isna().sum()
```

## Check For Null values using Visualization

```python
sns.heatmap(df.isnull())
```

## Sales Analysis based on Regions

```python
df.groupby("Region")['Sales'].sum().plot.bar()
plt.xlabel("Region")
plt.ylabel("Sales")
```

```python
df.groupby("Region")['Sales'].sum().plot.pie(autopct="%1.0f%%")
```

## Profit Analysis based on Regions

```python
df.groupby("Region")['Profit'].sum().plot.bar()
plt.xlabel("Region")
plt.ylabel("Profit")
```

```python
df.groupby("Region")['Profit'].sum().plot.pie(autopct="%1.0f%%")
```

## Sales analysis based on Category

```python
df.groupby("Category")['Sales'].sum().plot.bar()
```

```python
df.groupby("Category")['Sales'].sum().plot.pie(autopct="%1.0f%%")
```

```python
df.groupby("Category")['Profit'].sum().plot.pie(autopct="%1.0f%%")
```

## Analyse the Sales Statewise

```python
df.groupby("State")['Sales'].sum().plot.bar()
```

```python
df.groupby("State")['Profit'].sum().plot.bar()
```
